```python
# Define a base class "Animal"
class Animal:

    # Constructor method to initialize name and species attributes
    def __init__(self, name, species):
        self.name = name
        self.species = species

    # Method to display details of the animal
    def show_details(self):
        print(f"Name: {self.name}")
        print(f"Species: {self.species}")

# Define a derived class "Dog" that inherits from the "Animal" class
class Dog(Animal):
   
    # Constructor method to initialize name, species, and breed attributes
    def __init__(self, name, species, breed):
        super().__init__(name, species)  # Use super() to call the parent class constructor
        self.breed = breed
    
    # Method to display details of the dog, including the breed
    def show_details(self):
        super().show_details()  # Use super() to call the parent class's show_details method
        print(f"Breed: {self.breed}")

# Define a derived class "GoldenRetriever" that inherits from the "Dog" class
class GoldenRetriever(Dog):

    # Constructor method to initialize name, species, breed, and color attributes
    def __init__(self, name, species, breed, color):
        super().__init__(name, species, breed)  # Use super() to call the parent class constructor
        self.color = color

    # Method to display details of the Golden Retriever, including the color
    def show_details(self):
        super().show_details()  # Use super() to call the parent class's show_details method
        print(f"Color: {self.color}")

# Create a GoldenRetriever object with specified attributes
obj = GoldenRetriever("Tommy", "Dog", "Golden Retriever", "Black")

# Display details of the GoldenRetriever object
obj.show_details()
```

```python
# Define a base class "Vehicle"
class Vehicle:

    # Constructor method to initialize make and model attributes
    def __init__(self, make, model):
        self.make = make
        self.model = model

    # Method to display details of the vehicle
    def show_details(self):
        print(f"Make: {self.make}")
        print(f"Model: {self.model}")

    # Method to start the vehicle
    def start(self):
        print(f"{self.make} {self.model} is starting.")

    # Method to stop the vehicle
    def stop(self):
        print(f"{self.make} {self.model} is stopping.")

# Define a derived class "Car" that inherits from the "Vehicle" class
class Car(Vehicle):

    # Constructor method to initialize make, model, and fuel_type attributes
    def __init__(self, make, model, fuel_type):
        super().__init__(make, model)  # Use super() to call the parent class constructor
        self.fuel_type = fuel_type

    # Method to display details of the car, including the fuel type
    def show_details(self):
        super().show_details()  # Use super() to call the parent class's show_details method
        print(f"Fuel Type: {self.fuel_type}")

    # Method to honk the car's horn
    def honk(self):
        print(f"{self.make} {self.model} is honking its horn.")

# Define a derived class "ElectricCar" that inherits from the "Car" class
class ElectricCar(Car):

    # Constructor method to initialize make, model, fuel_type, and battery_capacity attributes
    def __init__(self, make, model, fuel_type, battery_capacity):
        super().__init__(make, model, fuel_type)  # Use super() to call the parent class constructor
        self.battery_capacity = battery_capacity

    # Method to display details of the electric car, including battery capacity
    def show_details(self):
        super().show_details()  # Use super() to call the parent class's show_details method
        print(f"Battery Capacity: {self.battery_capacity}")

    # Method to charge the electric car's battery
    def charge_battery(self):
        print(f"Charging the battery of {self.make} {self.model}.")

# Create an ElectricCar object with specified attributes
electric_car = ElectricCar("Tesla", "Model S", "Electric", "100 kWh")

# Display details of the ElectricCar object
electric_car.show_details()

# Start and stop the ElectricCar
electric_car.start()
electric_car.stop()

# Honk the horn of the ElectricCar
electric_car.honk()

# Charge the battery of the ElectricCar
electric_car.charge_battery()
```

```python
class DairyProduct:
    
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def product_details(self):
        print(f"Product: {self.name} - Price: {self.price} Rupees")

class Cheese(DairyProduct):
    
    def __init__(self, name, price, protein):
        super().__init__(name, price)
        self.protein = protein

    def nutrition_details(self):
        print(f"{self.name} contains {self.protein} grams of protein per serving")

class Butter(Cheese):
    
    def __init__(self, name, price, protein, expiry_date):
        super().__init__(name, price, protein)
        self.expiry_date = expiry_date

    def expiry_details(self):
        print(f"Expiry date: {self.expiry_date}")

# Create a Butter object
butter1 = Butter("Salted Butter", 220, 0.9, "5th August")
butter1.product_details()
butter1.nutrition_details()
butter1.expiry_details()

print("___________")

# Create another Butter object
butter2 = Butter("Unsalted Butter", 110, 1.5, "15th August")
butter2.product_details()
butter2.nutrition_details()
butter2.expiry_details()
```

```python
import tkinter as tk
from tkinter import messagebox

# Function to handle form submission
def submit_form():
    username = username_entry.get()
    password = password_entry.get()
    print("Username:", username)
    print("Password:", password)

# Create the main tkinter window
app = tk.Tk()
app.title("Registration Form")

# Create labels and entry fields for username and password
username_label = tk.Label(app, text="Username:")
username_label.pack()

username_entry = tk.Entry(app)
username_entry.pack()

password_label = tk.Label(app, text="Password:")
password_label.pack()

password_entry = tk.Entry(app, show="*")  # Use show="*" to hide the password
password_entry.pack()

# Create a checkbox for "Not a Robot"
robot_checkbox = tk.Checkbutton(app, text="I am not a robot")
robot_checkbox.pack()

# Create a submit button and bind it to the submit_form function
submit_button = tk.Button(app, text="Submit", command=submit_form)
submit_button.pack()

# Start the tkinter main loop
app.mainloop()
```

```python
import tkinter as tk

def submit():
    username = username_entry.get()
    password = password_entry.get()
    not_a_robot = not_a_robot_var.get()
    
    if not_a_robot:
        print(f"Username: {username}")
        print(f"Password: {password}")
        print("Registration successful!")
    else:
        print("Please confirm that you are not a robot.")

# Create the main tkinter window
root = tk.Tk()
root.title("Registration Form")

# Create a frame for the form
form_frame = tk.Frame(root, padx=20, pady=20)
form_frame.pack()

# Create labels and entry fields for username and password
username_label = tk.Label(form_frame, text="Username:")
username_label.grid(row=0, column=0, sticky="w")
username_entry = tk.Entry(form_frame)
username_entry.grid(row=0, column=1)

password_label = tk.Label(form_frame, text="Password:")
password_label.grid(row=1, column=0, sticky="w")
password_entry = tk.Entry(form_frame, show="*")
password_entry.grid(row=1, column=1)

# Create a checkbox for "Not a Robot"
not_a_robot_var = tk.BooleanVar()
not_a_robot_checkbox = tk.Checkbutton(form_frame, text="I am not a robot", variable=not_a_robot_var)
not_a_robot_checkbox.grid(row=2, columnspan=2)

# Create a submit button
submit_button = tk.Button(form_frame, text="Submit", command=submit)
submit_button.grid(row=3, columnspan=2)

# Start the tkinter main loop
root.mainloop()
```

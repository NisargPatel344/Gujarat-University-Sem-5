```python
import tkinter as tk

# Function to handle form submission
def submit_form():
    username = username_entry.get()
    password = password_entry.get()

    # Open a text file in write mode and append the username and password
    with open("user_data.txt", "a") as file:
        file.write(f"Username: {username}\nPassword: {password}\n\n")

    # Print a success message in the console
    print("Data submitted successfully!")

# Create the main tkinter window
app = tk.Tk()
app.title("Registration Form")

# Create labels and entry fields for username and password
username_label = tk.Label(app, text="Username:")
username_label.pack(pady=5)  # Adding padding to the top of the label

username_entry = tk.Entry(app)
username_entry.pack(pady=5)  # Adding padding to the top of the entry field

password_label = tk.Label(app, text="Password:")
password_label.pack(pady=5)  # Adding padding to the top of the label

password_entry = tk.Entry(app, show="*")  # Use show="*" to hide the password
password_entry.pack(pady=5)  # Adding padding to the top of the entry field

# Create a checkbox for "Not a Robot"
robot_checkbox = tk.Checkbutton(app, text="I am not a robot")
robot_checkbox.pack(pady=5)  # Adding padding to the top of the checkbox

# Create a submit button and bind it to the submit_form function
submit_button = tk.Button(app, text="Submit", command=submit_form, bg="green", fg="white")  # Adding background color and text color
submit_button.pack(pady=10)  # Adding more padding to the top and bottom of the button

# Start the tkinter main loop
app.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

def submit_form():
    username = username_entry.get()
    password = password_entry.get()
    
    if username and password:
        with open("user_data.txt", "a") as file:
            file.write(f"Username: {username}\nPassword: {password}\n\n")
        
        feedback_label.config(text="Data submitted successfully!", foreground="green")
        submit_button.config(state=tk.DISABLED)
    else:
        feedback_label.config(text="Please fill in both fields.", foreground="red")

app = tk.Tk()
app.title("Registration Form")

# Create a frame to organize elements
frame = ttk.Frame(app, padding=10)
frame.grid(column=0, row=0, sticky=(tk.W, tk.E))

# Username field
username_label = ttk.Label(frame, text="Username:")
username_label.grid(column=0, row=0, sticky=tk.W)

username_entry = ttk.Entry(frame)
username_entry.grid(column=1, row=0, sticky=(tk.W, tk.E))

# Password field
password_label = ttk.Label(frame, text="Password:")
password_label.grid(column=0, row=1, sticky=tk.W)

password_entry = ttk.Entry(frame, show="*")
password_entry.grid(column=1, row=1, sticky=(tk.W, tk.E))

# Checkbox
robot_checkbox = ttk.Checkbutton(frame, text="I am not a robot")
robot_checkbox.grid(columnspan=2, row=2, sticky=tk.W)

# Submit button
submit_button = ttk.Button(frame, text="Submit", command=submit_form)
submit_button.grid(columnspan=2, row=3)

# Feedback label
feedback_label = ttk.Label(frame, text="", foreground="green")
feedback_label.grid(columnspan=2, row=4)

app.mainloop()
```

```python
import tkinter as tk

# Function to handle form submission
def submit_form():
    student_name = name_entry.get()
    roll_number = roll_number_entry.get()
    student_id = student_id_entry.get()
    address = address_entry.get()
    contact = contact_entry.get()
    email = email_entry.get()

    # Open a text file in write mode and append the student data
    with open("studentData.txt", "a") as file:
        file.write(f"Student Name: {student_name}\n")
        file.write(f"Roll Number: {roll_number}\n")
        file.write(f"Student ID: {student_id}\n")
        file.write(f"Address: {address}\n")
        file.write(f"Contact: {contact}\n")
        file.write(f"Email: {email}\n")
        file.write("\n")

    # Print a success message in the console
    print("Data submitted successfully!")

# Create the main tkinter window
app = tk.Tk()
app.title("Student Registration Form")

# Create labels and entry fields for student information in a grid layout
tk.Label(app, text="Student Name:").grid(row=0, column=0, sticky="w", padx=10, pady=5)
name_entry = tk.Entry(app)
name_entry.grid(row=0, column=1, padx=10, pady=5)

tk.Label(app, text="Roll Number:").grid(row=1, column=0, sticky="w", padx=10, pady=5)
roll_number_entry = tk.Entry(app)
roll_number_entry.grid(row=1, column=1, padx=10, pady=5)

tk.Label(app, text="Student ID:").grid(row=2, column=0, sticky="w", padx=10, pady=5)
student_id_entry = tk.Entry(app)
student_id_entry.grid(row=2, column=1, padx=10, pady=5)

tk.Label(app, text="Address:").grid(row=3, column=0, sticky="w", padx=10, pady=5)
address_entry = tk.Entry(app)
address_entry.grid(row=3, column=1, padx=10, pady=5)

tk.Label(app, text="Contact:").grid(row=4, column=0, sticky="w", padx=10, pady=5)
contact_entry = tk.Entry(app)
contact_entry.grid(row=4, column=1, padx=10, pady=5)

tk.Label(app, text="Email:").grid(row=5, column=0, sticky="w", padx=10, pady=5)
email_entry = tk.Entry(app)
email_entry.grid(row=5, column=1, padx=10, pady=5)

# Create a submit button and place it in a grid layout
submit_button = tk.Button(app, text="Submit", command=submit_form, bg="green", fg="white")
submit_button.grid(row=6, columnspan=2, pady=10)

# Start the tkinter main loop
app.mainloop()
```

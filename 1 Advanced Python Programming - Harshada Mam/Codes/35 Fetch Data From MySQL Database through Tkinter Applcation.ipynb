```python
from tkinter import *
import mysql.connector
from tkinter import messagebox

def fetch_data():
    id = id_entry.get()
    name = name_entry.get()
    age = age_entry.get()
    
    try:
        conn = mysql.connector.connect(
            host="localhost",
            user="root",
            password="11111111",
            database="demo23"
        )

        cursor = conn.cursor()
        query = "SELECT * FROM Customer WHERE id = %s AND name = %s AND age = %s"
        cursor.execute(query, (id, name, age))
        
        user = cursor.fetchone()
        
        if user:
            messagebox.showinfo("Login Successful", "Welcome " + name)
            
        else:
            messagebox.showerror("Login Failed", "Invalid id, name, and age")

    except mysql.connector.Error as err:
        messagebox.showerror("Database Error", "Error connecting to the database: " + str(err))            

    finally:
        if conn.is_connected():
            cursor.close()
            conn.close()
            
root = Tk()
root.title("Login for Customer")

# Create a frame for data retrieval
retrieve_frame = Frame(root, padx=20, pady=20)
retrieve_frame.pack()

id_label = Label(retrieve_frame, text="ID:")
id_label.pack()
id_entry = Entry(retrieve_frame)
id_entry.pack()

name_label = Label(retrieve_frame, text="Name:")
name_label.pack()
name_entry = Entry(retrieve_frame)
name_entry.pack()

age_label = Label(retrieve_frame, text="Age:")
age_label.pack()
age_entry = Entry(retrieve_frame)
age_entry.pack()

login_button = Button(retrieve_frame, text="Login", command=fetch_data)
login_button.pack()

root.mainloop()
```

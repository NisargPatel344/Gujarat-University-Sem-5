```python
from tkinter import *
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="11111111",
    database="demo27"
)

cursor = conn.cursor()

def insert_data():
    id = id_entry.get()
    name = name_entry.get()
    age = age_entry.get()
    
    if id and name and age:
        sql = "INSERT INTO Customer (id,name,age) VALUES (%s,%s,%s)"
        val = (id,name,age)
        cursor.execute(sql,val)
        conn.commit()
        status_label.config(text="Data Inserted Successfully!")
        id_entry.delete(0, END)
        name_entry.delete(0, END)
        age_entry.delete(0, END)
        
    else:
        status_label.config(text="Please fill all fields")

root = Tk()
root.title("Insert Data into MySQL Database")

# Create a frame for data insertion
insert_frame = Frame(root, padx=20, pady=20)
insert_frame.pack()

id_label = Label(insert_frame, text="ID:")
id_label.pack()
id_entry = Entry(insert_frame)
id_entry.pack()

name_label = Label(insert_frame, text="Name:")
name_label.pack()
name_entry = Entry(insert_frame)
name_entry.pack()

age_label = Label(insert_frame, text="Age:")
age_label.pack()
age_entry = Entry(insert_frame)
age_entry.pack()

insert_button = Button(insert_frame, text="Insert Data", command=insert_data)
insert_button.pack()

status_label = Label(insert_frame, text="")
status_label.pack()

root.mainloop()

conn.close()
```

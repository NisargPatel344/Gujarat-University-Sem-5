```python
import mysql.connector
from tkinter import *
from tkinter import messagebox

# Create a database and table if they don't exist
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="11111111"
)

cursor = conn.cursor()

# Create the database if it doesn't exist
cursor.execute("CREATE DATABASE IF NOT EXISTS registration_form")

# Switch to the database
cursor.execute("USE registration_form")

# Create the 'users' table if it doesn't exist
cursor.execute("""
    CREATE TABLE IF NOT EXISTS users (
        id INT AUTO_INCREMENT PRIMARY KEY,
        first_name VARCHAR(255),
        middle_name VARCHAR(255),
        last_name VARCHAR(255),
        birth_date DATE,
        gender VARCHAR(10),
        country VARCHAR(255),
        phone VARCHAR(20),
        email VARCHAR(255),
        street VARCHAR(255),
        city VARCHAR(255),
        state VARCHAR(255),
        postal VARCHAR(20)
    )
""")
conn.commit()
conn.close()

# Function to insert data into the database
def insert_data():
    first_name = first_name_entry.get()
    middle_name = middle_name_entry.get()
    last_name = last_name_entry.get()
    birth_date = birth_date_entry.get()
    gender = gender_entry.get()
    country = country_entry.get()
    phone = phone_entry.get()
    email = email_entry.get()
    street = street_entry.get()
    city = city_entry.get()
    state = state_entry.get()
    postal = postal_entry.get()

    if any([
        first_name == "",
        last_name == "",
        birth_date == "",
        gender == "",
        country == "",
        phone == "",
        email == "",
        street == "",
        city == "",
        state == "",
        postal == ""
    ]):
        messagebox.showerror("Error", "Please fill in all fields")
        return

    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="11111111",
        database="registration_form"
    )
    cursor = conn.cursor()

    insert_query = """
        INSERT INTO users (
            first_name, middle_name, last_name, birth_date, gender, 
            country, phone, email, street, city, state, postal
        ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    data = (
        first_name, middle_name, last_name, birth_date, gender,
        country, phone, email, street, city, state, postal
    )

    cursor.execute(insert_query, data)
    conn.commit()
    conn.close()

    messagebox.showinfo("Success", "Data inserted successfully")

# Function to fetch data based on the name
def fetch_data():
    name = search_name_entry.get()
    
    if name == "":
        messagebox.showerror("Error", "Please enter a name to search")
        return

    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="11111111",
        database="registration_form"
    )
    cursor = conn.cursor()

    cursor.execute("SELECT * FROM users WHERE first_name = %s", (name,))
    result = cursor.fetchall()

    if not result:
        messagebox.showinfo("Info", "No data found")
    else:
        data_text.delete(1.0, END)
        for row in result:
            data_text.insert(END, f"ID: {row[0]}\nFirst Name: {row[1]}\nMiddle Name: {row[2]}\nLast Name: {row[3]}\nBirth Date: {row[4]}\nGender: {row[5]}\nCountry: {row[6]}\nPhone: {row[7]}\nEmail: {row[8]}\nStreet: {row[9]}\nCity: {row[10]}\nState: {row[11]}\nPostal: {row[12]}\n\n")

    conn.close()

# Create the main GUI window
root = Tk()
root.title("Registration Form")
root.geometry("600x600")  # Set the initial window size

# Create a background color
root.configure(bg="#f0f0f0")

# Labels and Entry fields for inserting data
first_name_label = Label(root, text="First Name:", bg="#f0f0f0")
first_name_label.pack()
first_name_entry = Entry(root)
first_name_entry.pack()

middle_name_label = Label(root, text="Middle Name:", bg="#f0f0f0")
middle_name_label.pack()
middle_name_entry = Entry(root)
middle_name_entry.pack()

last_name_label = Label(root, text="Last Name:", bg="#f0f0f0")
last_name_label.pack()
last_name_entry = Entry(root)
last_name_entry.pack()

birth_date_label = Label(root, text="Birth Date (YYYY-MM-DD):", bg="#f0f0f0")
birth_date_label.pack()
birth_date_entry = Entry(root)
birth_date_entry.pack()

gender_label = Label(root, text="Gender:", bg="#f0f0f0")
gender_label.pack()
gender_entry = Entry(root)
gender_entry.pack()

country_label = Label(root, text="Country:", bg="#f0f0f0")
country_label.pack()
country_entry = Entry(root)
country_entry.pack()

phone_label = Label(root, text="Phone:", bg="#f0f0f0")
phone_label.pack()
phone_entry = Entry(root)
phone_entry.pack()

email_label = Label(root, text="Email:", bg="#f0f0f0")
email_label.pack()
email_entry = Entry(root)
email_entry.pack()

street_label = Label(root, text="Street:", bg="#f0f0f0")
street_label.pack()
street_entry = Entry(root)
street_entry.pack()

city_label = Label(root, text="City:", bg="#f0f0f0")
city_label.pack()
city_entry = Entry(root)
city_entry.pack()

state_label = Label(root, text="State:", bg="#f0f0f0")
state_label.pack()
state_entry = Entry(root)
state_entry.pack()

postal_label = Label(root, text="Postal:", bg="#f0f0f0")
postal_label.pack()
postal_entry = Entry(root)
postal_entry.pack()

insert_button = Button(root, text="Insert Data", command=insert_data, bg="#4caf50", fg="white")
insert_button.pack()

# Labels and Entry field for fetching data
search_name_label = Label(root, text="Search by First Name:", bg="#f0f0f0")
search_name_label.pack()
search_name_entry = Entry(root)
search_name_entry.pack()

fetch_button = Button(root, text="Fetch Data", command=fetch_data, bg="#2196F3", fg="white")
fetch_button.pack()

# Text widget to display fetched data
data_text = Text(root, width=40, height=10)
data_text.pack()

root.mainloop()
```

```python
import pandas as pd
from sqlalchemy import create_engine, text, select
```

```python
# Define the database connection parameters
db_username = 'root'
db_password = '11111111'
db_host = 'localhost'  # Change to your MySQL server's hostname or IP address
db_port = '3306'       # Change to your MySQL server's port if necessary
db_database = 'company'
```

```python
# Create a SQLAlchemy engine to connect to the database
engine = create_engine(f'mysql+mysqlconnector://{db_username}:{db_password}@{db_host}:{db_port}/{db_database}')
```

```python
# Create a SQLAlchemy select statement
jobs = text("SELECT * FROM Jobs")
```

```python
# Execute the select statement and fetch the data into a DataFrame
with engine.connect() as connection:
    result = connection.execute(jobs)
    Jobs = pd.DataFrame(result.fetchall(), columns=result.keys())
```

```python
# Join Employees and Jobs on the 'Job_ID' column (Inner Join)
employee_job_inner = pd.merge(Employees, Jobs, left_on='Job_ID', right_on='Job_ID', how='inner')
```

```python
# Left join Employees and Jobs on the 'Job_ID' column
employee_job_left = pd.merge(Employees, Jobs, left_on='Job_ID', right_on='Job_ID', how='left')
```

```python
# Right join Employees and Jobs on the 'Job_ID' column
employee_job_right = pd.merge(Employees, Jobs, left_on='Job_ID', right_on='Job_ID', how='right')
```

```python
# Full outer join Employees and Jobs on the 'Job_ID' column
employee_job_outer = pd.merge(Employees, Jobs, left_on='Job_ID', right_on='Job_ID', how='outer')
```

```python
# First, join Employees and Jobs
employee_job = pd.merge(Employees, Jobs, left_on='Job_ID', right_on='Job_ID', how='inner')

# Then, join the result with Department
employee_job_department = pd.merge(employee_job, Department, left_on='Department_ID', right_on='Department_ID', how='inner')
```

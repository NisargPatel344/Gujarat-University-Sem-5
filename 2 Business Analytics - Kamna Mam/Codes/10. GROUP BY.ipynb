# Sales Table (2 Table)

```python
cnn = sqlite3.connect('data4.db')
cur = cnn.cursor()
```

```python
cur.execute('''
CREATE TABLE Sales (
    Product_ID INT,
    Sell_Price FLOAT,
    Quantity INT,
    State VARCHAR(20)
);
''')
```

```python
cur.execute('''
INSERT INTO Sales VALUES
(121, 320.0, 3, 'California'),
(121, 320.0, 6, 'Texas'),
(121, 320.0, 4, 'Alaska'),
(123, 290.0, 2, 'Texas'),
(123, 290.0, 7, 'California'),
(123, 290.0, 4, 'Washington'),
(121, 320.0, 7, 'Ohio'),
(121, 320.0, 2, 'Arizoa'),
(123, 290.0, 8, 'Colorado');
''')
```

```python
cur.execute('''
CREATE TABLE Products (
    Product_ID INT,
    Cost_Price FLOAT
);
''')
```

```python
cur.execute('''
INSERT INTO Products VALUES
(121, 270.0),
(123, 250.0);
''')
```

```python
cur.execute('''
SELECT * FROM Products;
''')

result = cur.fetchall()

df = pd.DataFrame(result, columns=[item[0] for item in cur.description])
display(df)

cur.execute('''
SELECT * FROM Sales;
''')

result = cur.fetchall()

df = pd.DataFrame(result, columns=[item[0] for item in cur.description])
display(df)
```

```python
cur.execute('''
SELECT Product_ID, SUM(Sell_Price * Quantity) AS Revenue
FROM Sales
GROUP BY Product_ID;
''')

result = cur.fetchall()

df = pd.DataFrame(result, columns=[item[0] for item in cur.description])
display(df)
```

```python
cur.execute('''
SELECT S.Product_ID, SUM((S.Sell_Price - P.Cost_Price) * S.Quantity) AS Profit
FROM Sales 
AS S JOIN Products AS P ON S.Product_ID = P.Product_ID
GROUP BY S.Product_ID
''')

result = cur.fetchall()

df = pd.DataFrame(result, columns=[item[0] for item in cur.description])
display(df)
```

```python
cur.execute('''
SELECT S.Product_ID, SUM((S.Sell_Price - P.Cost_Price) * S.Quantity) AS Profit
FROM Sales 
AS S INNER JOIN Products AS P 
WHERE S.Product_ID = P.Product_ID
GROUP BY P.Product_ID
''')

result = cur.fetchall()

df = pd.DataFrame(result, columns=[item[0] for item in cur.description])
display(df)
```

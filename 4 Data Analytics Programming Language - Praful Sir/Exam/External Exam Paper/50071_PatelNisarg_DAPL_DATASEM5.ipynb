### 3.	Perform the Bivariate Analysis on Given dataset,Use appropriate visualizations to explore relationships between variables and draw meaningful conclusions.(DATA2.CSV)(10 Marks)


```python
# Import the NumPy library for numerical operations
import numpy as np

# Import the Pandas library for data manipulation and analysis
import pandas as pd

# Import the Seaborn library for data visualization
import seaborn as sns

# Import the Matplotlib library for creating static, animated, or interactive visualizations
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('DATA2.csv')
```

```python
df.head()
```

```python
# Display the information of the data
df.info()
```

```python
# Display the shape of the data
df.shape
```

```python
# Display the columns of the data
df.columns
```

```python
# Display the count, mean, std, minimum, Q1, Q2, Q3, and maximum
df.describe().T
```

```python
# Display the histograms of all the numerical data in the dataset
df.hist(figsize=(12, 10))
plt.suptitle('Histograms for Numerical Columns', y=1.02, fontsize=16)
plt.show()
```

```python
df.columns
```

```python
df1 = df.groupby('airport')['price'].sum().reset_index().sort_values(by='price', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x='airport', y='price', data=df1, color='green')
plt.xlabel('airport')
plt.ylabel('price')
plt.xticks(rotation=90, ha='right')
plt.show()
```

**Inference**
- The price of the resedentail area near airport are very high
- The price of the residentail area not near airport are very low

```python
df1 = df.groupby('waterbody')['price'].sum().reset_index().sort_values(by='price', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x='waterbody', y='price', data=df1, color='green')
plt.xlabel('waterbody')
plt.ylabel('price')
plt.xticks(rotation=90, ha='right')
plt.show()
```

**Inference**
- The price of the resedentail area near river are very high
- The price of the residentail area near lake are very low

```python
df1 = df.groupby('airport')['age'].mean().reset_index().sort_values(by='age', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x='airport', y='age', data=df1, color='blue')
plt.xlabel('airport')
plt.ylabel('age')
plt.xticks(rotation=90, ha='right')
plt.show()
```

**Infernce**
- The average age of the residenital near airport and not nearairport are almost same

```python
df1 = df.groupby('waterbody')['age'].mean().reset_index().sort_values(by='age', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x='waterbody', y='age', data=df1, color='blue')
plt.xlabel('waterbody')
plt.ylabel('age')
plt.xticks(rotation=90, ha='right')
plt.show()
```

**Inference**
- The average age of the resedentail area near none are very high
- The average age of the residentail area near river are very low

```python
df1 = df.groupby('airport')['Sold'].count().reset_index().sort_values(by='Sold', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x='airport', y='Sold', data=df1, color='pink')
plt.xlabel('airport')
plt.ylabel('Sold')
plt.xticks(rotation=90, ha='right')
plt.show()
```

**Inference**
- The number of counts of the resedentail area near airport are very high
- The number of counts of the residentail area not near airport are very low

```python
df1 = df.groupby('waterbody')['Sold'].count().reset_index().sort_values(by='Sold', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x='waterbody', y='Sold', data=df1, color='pink')
plt.xlabel('waterbody')
plt.ylabel('Sold')
plt.xticks(rotation=90, ha='right')
plt.show()
```

**Inference**
- The number of counts of the resedentail area near river are very high
- The number of counts of the residentail area near lake are very low

### Concat

```python
# Step 1 :- Create two dataframes

import numpy as np
import pandas as pd

# First Dataframe
df_1 = pd.DataFrame(
    {
        "Name":["Ram","Shyam","Ham","Mahesh","Dan",np.nan],
        "Marks":[15,20,15,20,18,17],
        "Branch":["Mobile","NS","Fintech","BIA","Data","Game"],
    }
)

# Second Dataframe
df_2 = pd.DataFrame(
    {
        "Name":["Virat","Rohit","Ashwin",np.nan,np.nan,"Aayush"],
        "Marks":[20,22,18,np.nan,22,14],
        "Branch":["Data","Game","Mobile","NS","Fintech","BIA"],
    }
)
```

```python
# Show the First Dataframes :

df_1.head()
```

```python
# Show the Second Dataframes :

df_2.head()
```

```python
# Concatinate both the dataframes

df_concat = pd.concat([df_1,df_2])
df_concat
```

```python
# Column Wise Cocatination
# Concatinate both the dataframes

df_concat = pd.concat([df_1,df_2],axis = 1)
df_concat
```

```python
# Column Wise Cocatination
# Concatinate both the dataframes

df_concat = pd.concat([df_1,df_2],axis = 0,ignore_index=True)
df_concat
```

# Bivariate Analysis

```python
# Sort DataFrame by Total Sales
df_sales_sorted = df.groupby('State')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)

# Total Sales by State (Bar chart)
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Sales', data=df_sales_sorted, color='blue')
plt.title('Total Sales by State (Descending Order)')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
# Sort DataFrame by Total Profit
df_profit_sorted = df.groupby('State')['Profit'].sum().reset_index().sort_values(by='Profit', ascending=False)

# Total Profit by State (Bar chart)
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Profit', data=df_profit_sorted, color='green')
plt.title('Total Profit by State (Descending Order)')
plt.xlabel('State')
plt.ylabel('Total Profit')
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
# Sort DataFrame by Total Sales
df_sales_by_region_sorted = df.groupby('Region')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)

# Total Sales by Region (Bar chart)
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Sales', data=df_sales_by_region_sorted, color='blue')
plt.title('Total Sales by Region (Descending Order)')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```
